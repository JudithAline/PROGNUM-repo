```python
# create 2 sets, with names night1 and night2
night1 = {1, 3, 2, 4, 0, 3, 7, 5, 2, 9, 2, 4, 3}
night2 = {0, 0, 2, 4, 2, 4, 3, 2, 1, 1, 1, 2, 0}

inters = set.intersection(night1, night2)
b = list(inters)
print(b)

if 5 in list(inters):
    print('5 is an element of the intersection list')
if 5 not in list(inters):
    print('5 is not an element of the intersection list')

if 5 in set(night1):
    print('5 is a member of the set version of night1')
if 5 not in set(night1):
    print('5 is not a member of the set version of night1')
```
