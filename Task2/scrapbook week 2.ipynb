```python
# 2.16 create 2 sets, named night1 and night2
night1 = {1, 3, 2, 4, 0, 3, 7, 5, 2, 9, 2, 4, 3}
night2 = {0, 0, 2, 4, 2, 4, 3, 2, 1, 1, 1, 2, 0}

inters = set.intersection(night1, night2)   # intersection of set night1 and night2
a = list(inters)                            
print(a)

differ = set.difference(night1, night2)     # difference between set night1 and night2
b = list(differ)                            
print(b)

union = set.union(night1, night2)           # combine set night1 and night2 without duplicates
c = list(union)                             
print(c)

night1 = set(night1 & night2)              # make a new set with only intersections
d = list(night1)
print(d)
```
